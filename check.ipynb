```python
from diffusers import StableDiffusionImg2ImgPipeline
import torch
from PIL import Image

# Load the model
pipe = StableDiffusionImg2ImgPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float32
).to("cpu")

# Load an input image
init_image = Image.open("./generated_image.png").convert("RGB")
init_image = init_image.resize((512, 512))

# Prompt and strength (how much to change from original)
prompt = "sun in the sky shining on the objects"
strength = 0.5  # 0.0 = no change, 1.0 = complete transformation
guidance_scale = 0.5

# Generate image
output = pipe(prompt=prompt, image=init_image, strength=strength, guidance_scale=guidance_scale)
output_image = output.images[0]

# Save or display the result
output_image.save("output_image.png")
```
